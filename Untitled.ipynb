```python
totalBlocks = 0
ind = 0
allCount = 0
compCount = 0
numBlocksWithKind = 0
procedureReturns = []
kindsDict = {}
totalNumBlocksBesidesGlobalDecls = 0
totalNumCompBlocksWOGlobalDecls = 0
declTypeKinds = {}
numComponentBlocksWithoutKind = 0
nonComponentBlockTypes = {}
nonComponentBlockTypesKinds = {}
topBlocksWithNoComps = 0

def zeroOut():
    global totalBlocks, ind, allCount, compCount, numBlocksWithKind, procedureReturns
    global kindsDict, totalNumBlocksBesidesGlobalDecls
    global totalNumCompBlocksWOGlobalDecls, declTypeKinds, numComponentBlocksWithoutKind
    global nonComponentBlockTypes, nonComponentBlockTypesKinds, topBlocksWithNoComps
    totalBlocks = 0
    ind = 0
    allCount = 0
    compCount = 0
    numBlocksWithKind = 0
    procedureReturns = []
    kindsDict = {}
    totalNumBlocksBesidesGlobalDecls = 0
    totalNumCompBlocksWOGlobalDecls = 0
    declTypeKinds = {}
    numComponentBlocksWithoutKind = 0
    nonComponentBlockTypes = {}
    nonComponentTypesKinds = {}
    topBlocksWithNoComps = 0

def enumerateBlockInfo(eq, blk):
    global totalBlocks, ind, allCount, compCount, numBlocksWithKind, procedureReturns, kindsDict
    global totalNumBlocksBesidesGlobalDecls
    global totalNumCompBlocksWOGlobalDecls, declTypeKinds, numComponentBlocksWithoutKind
    global nonComponentBlockTypes, nonComponentBlockTypesKinds, topBlocksWithNoComps
    totalBlocks += 1
    ind += 1
    tmpAll = ab.countAllBlocks(blk)
    tmpComp, tmpGeneric = ab.countComponents(blk)
    allCount += tmpAll
    compCount += tmpComp
    if 'kind' in blk:
        numBlocksWithKind += 1
        k = blk['kind']
        tipe = blk['*type']
        if tipe == "procedures_defreturn":
            procedureReturns.append(eq.projectName + "-" + eq.programmerName)
        if k in kindsDict:
            kindsDict[k] += 1
        else:
            kindsDict[k] = 1
        if k == "declaration":
            totalNumBlocksBesidesGlobalDecls += tmpAll
            totalNumCompBlocksWOGlobalDecls += tmpComp
            if k in declTypeKinds:
                if tipe in declTypeKinds[k]:
                    declTypeKinds[k][tipe]['num'] += 1
                    declTypeKinds[k][tipe]['all'] += tmpAll
                    declTypeKinds[k][tipe]['comp'] += tmpComp
                    declTypeKinds[k][tipe]['generic'] += tmpGeneric
                else:
                    declTypeKinds[k][tipe] = {}
                    declTypeKinds[k][tipe]['num'] = 1
                    declTypeKinds[k][tipe]['all'] = tmpAll
                    declTypeKinds[k][tipe]['comp'] = tmpComp
                    declTypeKinds[k][tipe]['generic'] = tmpGeneric
            else:
                declTypeKinds[k] = {}
                declTypeKinds[k][tipe] = {}
                declTypeKinds[k][tipe]['num'] = 1
                declTypeKinds[k][tipe]['all'] = tmpAll
                declTypeKinds[k][tipe]['comp'] = tmpComp
                declTypeKinds[k][tipe]['generic'] = tmpGeneric
    else:
        numComponentBlocksWithoutKind+=1

    tipe = blk['*type']
    if (tmpComp == 0 and tmpAll != 0) or (tipe == "component_event" and tmpComp == 1):
        k = blk['kind']

        if tipe not in nonComponentBlockTypes:
            nonComponentBlockTypes[tipe] = 1
        else:
            nonComponentBlockTypes[tipe] += 1
        if k in nonComponentBlockTypesKinds:
            if tipe in nonComponentBlockTypesKinds[k]:
                nonComponentBlockTypesKinds[k][tipe] += 1
            else:
                nonComponentBlockTypesKinds[k][tipe] = 1
        else:
            nonComponentBlockTypesKinds[k] = {}
            nonComponentBlockTypesKinds[k][tipe] = 1
            topBlocksWithNoComps += 1
            #print "Project " + eq.projectName + "
            #by " + eq.programmerName + " has no component blocks"
    if ind % 1000 == 0:
        print allCount, compCount
```

```python
users = {}

def recordUsersAndProjects(eq, blk):
    global users
    if eq.programmerName not in users:
        users[eq.programmerName] = []
    if eq.projectName not in users[eq.programmerName]:
        users[eq.programmerName].append(eq.projectName)
```

```python
users = {}
ab.iterateOverProjectSets(equivs, recordUsersAndProjects)
```

```python
handlersDict = {"proj": [], 'user': [], 'numBlocks': [], 'type': []}
def recordUsersHandlers(eq, blk):
    if blk['kind'] == "declaration":
        global handlersDict
        handlersDict["proj"].append(eq.projectName)
        handlersDict["user"].append(eq.programmerName)
        handlersDict["numBlocks"].append(ab.countAllBlocks(blk))
        handlersDict["type"].append(blk["*type"])
```
